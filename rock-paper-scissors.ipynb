```python
import random

user_wins = 0
computer_wins = 0

choices = ["rock", "paper", "scissors"]

while True:
    user_choice = input("Type your selection 'rock, paper, scissors' or type Q to quit. ").lower()
   
    if user_choice == "q" :
        break
    if user_choice not in choices:
        print("Please type a valid value! \n")
        continue
        
    print("Your choice", user_choice)
    
    
    random_number = random.randint(0,2)
    #0 for rock, 1 for paper, 2 for scissors
    computer_choice = choices[random_number]
    print("Computer choice", computer_choice)
    
    if user_choice == "rock" and computer_choice == "scissors":
        print("You win! \n")
        user_wins += 1
        
    elif user_choice == "paper" and computer_choice == "rock":
        print("You win! \n")
        user_wins += 1        
        
    elif user_choice == "scissors" and computer_choice == "paper":
        print("You win! \n")
        user_wins += 1
        
    elif user_choice == "scissors" and computer_choice == "scissors":
        print("Draw! \n")
    
    elif user_choice == "paper" and computer_choice == "paper":
        print("Draw! \n")
    
    elif user_choice == "rock" and computer_choice == "rock":
        print("Draw! \n")
       
    else:
        print("Computer wins! \n")
        computer_wins += 1
        
print("Game is over! You won ", user_wins, "times. Computer won ", computer_wins, "times.")

        
    
```
